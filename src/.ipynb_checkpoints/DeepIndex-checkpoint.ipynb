# Tree Finding

The algorithm aims to optimise a search tree via stochastic optimization.

## Subproblem

In this section, we aim to optimise a classification tree. 

The loss function:
$$Loss(\theta) = -\sum_{i \in [N]}\sum_{j \in [k]}\text{log}\left(p_\theta^{(i,j)}\right)$$
where $k$ is the number of classes, and $\sum_{i \in [k]}p_\theta^{(i)} = 1$

We solve it by changing $\theta$ to a random variable and optimising with the expected loss with respect to the parameters its distribution.

The structure of our model is a tree:
- split 
    - split 
        - c
        - d
    - split
        - e
        - f
    ...

In each node, the tree goes either left or right based on binarized linear function.

We optimise the parameters of the splits, which are some $\beta$ vector variable. We do that, via optimization of parameters of distribution of their randomized versions. Our actual loss function works as follows:
$$ L(\phi) = \mathbb{E}_{\theta \sim q_\phi} Loss(\theta) = \mathbb{E}_{\theta \sim q_\phi} Loss = -\sum_{i \in [N]} \sum_{j \in [k]}\mathbb{E}_{\theta \sim q_\phi}\text{log}\left(p_\theta^{(i,j)}\right) $$

For a tree:
$$\mathbb{E}_{\theta \sim q_\phi}\text{log}\left(p_\theta^j\right) = \mathbb{E}_{\theta \sim q_\phi}\text{log}\left(p_\theta(C = j)\right) =$$ 
$$ = \mathbb{E}_{\theta \sim q_\phi}\text{log}\left(p_\theta(C = j | L)p_\beta(L) + p_\theta(C = j | R)p_\beta(R) \right) $$

Given the distribution of $\beta$, we can calculate (or approximate) probabilities $p_\beta(L)$ and $p_\beta(R) = 1 - p_\beta(L)$ via a closed form solution. $p_\theta(C = j | L)$ and $p_\theta(C = j | R)$
